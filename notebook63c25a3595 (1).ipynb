```python
import pandas as pd
pd.plotting.register_matplotlib_converters()
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
```

```python
filepath= "../input/globalshala-week-1-assignment/GlobalShala Week 1 Assignment.csv"
df= pd.read_csv(filepath)
```

```python
df.head()
```

```python
plt.figure(figsize=(8,6))
plt.title(' Bar chart that shows how campaigns are performing')
#sns.barplot(y= df["Campaign ID"], x=df['Cost per Result (CPR)'])
df_new = df.rename(columns={'Cost per Result (CPR)': 'CPR',"Campaign ID":"Campaign_ID"})

sns.barplot(x='CPR',
            y="Campaign_ID", data=df_new,
            order=df_new.sort_values("CPR").Campaign_ID)
```

```python
plt.figure(figsize=(8,6))
plt.title(' Bar chart that shows how campaigns are performing')
df_new = df.rename(columns={'Cost Per Click (CPC)': 'CPC',"Campaign ID":"Campaign_ID"})
sns.barplot(x='CPC',
            y="Campaign_ID", data=df_new,
            order=df_new.sort_values("CPC").Campaign_ID)
```
